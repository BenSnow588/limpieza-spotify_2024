Veamos la correlacion : La correlación es una medida estadística que indica qué tan relacionada está una variable con otra. En términos simples, te dice si dos cosas tienen una conexión y en qué dirección.

🔹 Tipos de correlación: 1️⃣ Correlación positiva (> 0): Cuando una variable sube, la otra también tiende a subir.

Ejemplo: Más horas de estudio suelen llevar a mejores calificaciones.

2️⃣ Correlación negativa (< 0): Cuando una variable sube, la otra tiende a bajar.

Ejemplo: A mayor velocidad de un automóvil, menos tiempo se tarda en llegar a un destino.

3️⃣ Correlación nula (≈ 0): No hay una relación clara entre las variables.

Ejemplo: El color de tu camisa y la cantidad de lluvia en un día.

🔢 Cómo se interpreta en Pandas: El método df.corr() calcula la correlación entre todas las columnas numéricas en un DataFrame. Si ves un valor cercano a 1, hay una fuerte relación positiva. Si ves un valor cercano a -1, hay una fuerte relación negativa. Si ves un

```python
print(df.select_dtypes(include=['number']).corr())
```

vamos a ver la cantidad de registros por año 

```python
import io
import matplotlib.pyplot as plt
from IPython.display import Image, display

# Crea un buffer en memoria para guardar la imagen
buffer = io.BytesIO()

# Genera un gráfico de barras con la cantidad de registros por cada año
df["year"].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')

# Agrega etiquetas a los ejes
plt.xlabel("Año")
plt.ylabel("Cantidad de registros")
plt.title("Distribución de registros por año")

# Guarda el gráfico en el buffer en formato PNG
plt.savefig(buffer, format='png')

# Reinicia el puntero del buffer para poder leer la imagen desde el inicio
buffer.seek(0)

# Muestra la imagen en el notebook
display(Image(data=buffer.getvalue()))

```
